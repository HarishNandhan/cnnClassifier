```python
def get_product(x:int,y:int) -> int :
    return x * y
```

```python
get_product(x=2,y=4)
```

```python
#why we use ensure_annotations?

from ensure import ensure_annotations

@ensure_annotations
def get_product(x:int,y:int) -> int :
    return x * y
```

```python
get_product(x=19,y="Harish")
```
